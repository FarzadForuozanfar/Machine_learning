```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([13, 40, 65, 85, 100])
y = np.array([110, 138, 149, 174, 184])

font = {'family':'serif','color':'red','size':25}

plt.plot(x, y)
plt.title("BMI", fontdict = font)
plt.xlabel("weight of people")
plt.ylabel("height of peoplr")
plt.grid(color = 'black', linestyle = ':', linewidth = 0.8)
plt.show()
```

```python
#subplots

#plot 1:
x = np.array([1396, 1397, 1398, 1399, 1400])
y = np.array([510,484,123,475,147])


plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("Tehran")

#plot 2:
x = np.array([1396, 1397, 1398, 1399, 1400])
y = np.array([470, 370, 364, 305, 190])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("Mashhad")

plt.suptitle("Rainfall")
plt.show()
```

```python
#scatter

x = np.random.randint(100, size=(100))
y = x * np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 7 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='rainbow_r')

plt.colorbar()

plt.show()
```

```python
x = np.array(["Ali", "Behnam", "Sara", "Donya"])
y = np.array([20,18.5,12,14])

plt.bar(x, y, width = 0.1,color = "black")
plt.title("Math")
plt.show()
```

```python
#histograms

x = np.random.normal(100, 20, 140)

plt.hist(x, color = 'red')
plt.show()
```

```python
#pie chart

y = np.array([10,15,20])
mylabels = ["c++", "python", "CSS3"]

plt.pie(y, labels = mylabels, shadow = True)
plt.legend(title = "Project")
plt.show()
```
